```python
import openai
import os
import dotenv
from IPython.display import Markdown, display

# env 설정
dotenv_file = dotenv.find_dotenv()
dotenv.load_dotenv(dotenv_file)
# print(os.environ["OPENAI_API_KEY"])
openai_api_key = os.environ["OPENAI_API_KEY"]

from langchain.chains import ConversationChain
from langchain_openai import ChatOpenAI
```

```python
def print_markdown(text):
    display(Markdown(text))
```

```python
# txt파일 load
file_path = "C:/Users/scvts/Desktop/jiwon/prompt/text_data/treatment_local_balance/운남한의원.txt"

with open(file_path, 'r', encoding='utf-8') as file:
    input_content = file.read()

input_content = input_content.replace("\n", " ")
print(input_content)
```

```python
# 프롬프트
user_text_1 = f'''
# {input_content}
위의 대화는 환자와 의사의 상담 내용입니다. 주소증, 발병일, 과거력, 가족력, 현병력으로 나눠서 요약해 주세요. 
주소증은 환자가 주로 호소하는 증상으로 환자가 병원에 오게 된 주요 이유 입니다.
발병일은 환자의 증상이 시작된 날짜를 말합니다. 
과거력은 환자의 현재 건강상태에 영향을 미칠 가능성이 있는 과거의 병력을 말합니다.
가족력은 환자의 가족 구성원이 특정 질환을 갖고 있는 것을 말합니다.
현병력은 주소증과 관련되어 최근에 나타난 건강 변화를 의미합니다.

요약에 대한 근거도 제시해 줘야 합니다. 근거는 환자와 의사의 상담 내용에서 전부 발췌해야 합니다. 

결과의 예시는 다음과 같습니다.

[요약 결과]
주소증 : 지속적인 피로감
발병일 : 약 한 달 전
과거력 : 없음
가족력 : 없음
현병력 : 수면 부족, 불규칙적인 식사, 운동 부족

[근거]
주소증 : "요즘 몸이 계속 피곤해요"
발병일 : "대략 한 달 정도 된 것 같아요"
과거력 : 없음
가족력 : 없음
현병력 : "밤에 잠도 잘 안 와서 그런가 봐요", "보통 4~5시간 정도만 자요", "요즘은 밥을 제대로 먹지 못하고 있어요. 일이 너무 바빠서...", "운동은 거의 안 해요. 시간이 없어서..."
# '''
```

```python
prompt_save_text = "위의 대화는 환자와 의사의 상담 내용입니다." + user_text_1.split("위의 대화는 환자와 의사의 상담 내용입니다.")[1]
```

```python
# GPT-4 인스턴스 생성
llm = ChatOpenAI(openai_api_key=openai_api_key, temperature=0, model_name="gpt-3.5-turbo-0125")  # gpt-4-0125-preview, gpt-3.5-turbo-0125
conversation_with_gpt4p0 = ConversationChain(llm=llm)

result = conversation_with_gpt4p0.invoke(user_text_1)
```

```python
print_markdown(result['response'])
```

```python
prompt_save_text = "위의 대화는 환자와 의사의 상담 내용입니다." + user_text_1.split("위의 대화는 환자와 의사의 상담 내용입니다.")[1]
```

```python
# 요약 결과 + 프롬프트 저장
file_name = "C:/Users/scvts/Desktop/jiwon//prompt/result/gpt3p5/운남_요약_COT_1.txt"
with open(file_name, 'w', encoding="utf-8") as file:
    file.write(result['response'])
    file.write("\n\n=====prompt=====\n\n")
    file.write(prompt_save_text)
```
