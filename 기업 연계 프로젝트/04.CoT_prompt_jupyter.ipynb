```python
import openai
import os
import dotenv
from IPython.display import Markdown, display

# env 설정
dotenv_file = dotenv.find_dotenv()
dotenv.load_dotenv(dotenv_file)
# print(os.environ["OPENAI_API_KEY"])
openai_api_key = os.environ["OPENAI_API_KEY"]

from langchain.chains import ConversationChain
from langchain_openai import ChatOpenAI
from langchain.schema import HumanMessage, SystemMessage
```

```python
def print_markdown(text):
    display(Markdown(text))
```

```python
# txt파일 load
file_path = "C:/Users/scvts/Desktop/jiwon/prompt/text_data/treatment_local_balance/하슬라한의원.txt"

with open(file_path, 'r', encoding='utf-8') as file:
    input_content = file.read()

input_content = input_content.replace("\n", " ")
print(input_content)
```

```python
# 프롬프트
conversation_text = []

user_text_1 = f'''
# {input_content}

위의 대화는 환자와 의사의 상담 내용입니다. 주소증, 발병일, 과거력, 가족력, 현병력으로 나눠서 요약해 주세요. 아래과 같이 단계별로 생각해보세요. 

1. 환자가 병원에 왔습니다. 환자가 주로 호소하는 증상과 환자가 병원에 오게 된 주요 이유를 파악합니다. 이것이 바로 주소증입니다.
2. 환자에게 직접 물어보거나, 환자가 자연스럽게 불편함을 표현하도록 유도하여 주소증을 확인해보세요.
3. 주소증이 왜 환자가 병원에 오게 만들었는지 이해하는 것이 중요합니다. 주소증이 환자의 일상생활에 어떤 영향을 미치고 있는지 생각해 보세요.
4. 주소증은 환자 본인의 증상만 해당이 됩니다. 타인의 증상이 들어가면 안됩니다.
5. 주소증에는 병력만 들어가야 합니다.
6. 주소증에 해당 되는 것을 출력해 주세요.

# '''

user_text_2 = f'''
아래과 같이 단계별로 생각해보세요.

1. 환자의 증상이 언제 처음 발생했는지를 파악하는 것이 중요합니다. 이것이 바로 '발병일'입니다.
2. 환자에게 그 증상이 처음으로 나타난 날짜를 물어봅시다.
3. 환자가 정확한 날짜를 기억하지 못한다면, 대략적인 시기나 그때의 상황을 물어보는 것도 도움이 될 수 있습니다.
4. '발병일'을 알아내는 것은 증상의 원인을 찾거나, 진행 상태를 파악하는 데 중요한 단서가 됩니다.
5. '발병일'을 바탕으로 증상의 발병 패턴, 지속 시간, 심화 정도 등을 분석해봅시다.
6. 발병일에 해당 되는 것을 출력해 주세요.
'''

user_text_3 = f'''
아래과 같이 단계별로 생각해보세요.

1. 환자의 현재 건강 상태를 이해하는데, 환자의 과거 병력인 '과거력'이 중요한 역할을 합니다.
2. 환자에게 이전에 앓았던 병이나 질병, 그리고 그에 따른 치료 과정에 대해 물어봅시다.
3. 과거력이 환자 본인의 병력이 맞는지 확인하세요. 타인의 병력이 과거력에 들어가면 안됩니다.
4. '과거력'은 병의 원인을 파악하거나, 병의 진행을 이해하는데 중요한 정보가 될 수 있습니다.
5. '과거력'에는 병력만 들어가야 합니다.
6. 과거력에 해당되는 것을 출력해주세요.

'''

user_text_4 = f'''
아래과 같이 단계별로 생각해보세요.

1. 환자의 가족 중에서 질병을 앓고 있는 사람이 있는지 확인하는 것이 중요합니다. 이것이 바로 '가족력'입니다.
2. 환자에게 가족 중에 질병을 앓고 있는 사람이 있는지, 그리고 그 질병이 무엇인지 확인하세요.
3. 특히, 가족력이 환자의 가족에 해당하는 질병인지 한번 확인하세요. 가족력에는 환자의 가족이 아닌 사람이 들어가면 안된니다.
4. 가족력에 해당되는 것을 출력해주세요.

'''

user_text_5 = f'''
아래과 같이 단계별로 생각해보세요.

1. 환자의 현재 건강 상태를 이해하는데 있어서, 최근에 나타난 건강 변화인 '현병력'이 중요합니다.
2. 환자에게 주소증과 관련하여 최근에 어떤 건강 상태의 변화가 있었는지 물어봅시다.
3. 이 변화가 언제부터 시작되었는지, 그리고 변화의 정도는 어느 정도인지를 파악하는 것이 중요합니다.
'현병력'은 병의 진행 상태를 파악하고, 적절한 치료 방향을 설정하는데 중요한 정보가 될 수 있습니다.
4, 이렇게 환자의 '현병력'을 체계적으로 파악하면, 환자의 현재 상태를 더 정확하게 이해하고, 적절한 진단과 치료 방향을 모색하는 데 도움이 될 것입니다.
5. 현병력에는 병력만 들어가야 합니다.
6. 현병력에 해당되는 것을 출력해주세요.
'''

user_text_6 = f'''
위에서 나온 5가지 결과를 주소증, 발병일, 과거력, 가족력, 현병력으로 나눠서 "요약 결과"에 출력해주세요. "요약 결과"는 생략되는 부분없이 전부 출력해야 합니다. "요약 결과"는 단답형으로 출력되어야 합니다.
"요약 결과"에 대한 "근거"도 제시해 줘야 합니다. "근거"는 환자와 의사의 상담 내용에서 전부 발췌해야 합니다. "근거"는 상담내용의 내용 그대로를 보여줘야 합니다.

최종 "요약 결과"와 "근거" 예시는 아래와 같습니다.

[요약 결과]
주소증 : 지속적인 피로감, 두통
발병일 : 피로감(약 한 달 전), 두통(약 2주전)
과거력 : 없음
가족력 : 없음
현병력 : 수면 부족, 불규칙적인 식사, 운동 부족

[근거]
주소증 : "요즘 몸이 계속 피곤해요", "머리가 아픈데"
발병일 : "계속 피곤했던건 대략 한 달 정도 된 것 같아요", "머리가 2주정도 전부터 아팠어요."
과거력 : 없음
가족력 : 없음
현병력 : "밤에 잠도 잘 안 와서 그런가 봐요", "보통 4~5시간 정도만 자요", "요즘은 밥을 제대로 먹지 못하고 있어요. 일이 너무 바빠서...", "운동은 거의 안 해요. 시간이 없어서..."

'''

conversation_text.append(user_text_1)
conversation_text.append(user_text_2)
conversation_text.append(user_text_3)
conversation_text.append(user_text_4)
conversation_text.append(user_text_5)
conversation_text.append(user_text_6)
```

```python
llm = ChatOpenAI(openai_api_key=openai_api_key, temperature=0, model_name="gpt-3.5-turbo-0125")

conversation_with_gpt3p5 = ConversationChain(llm=llm)

for content in conversation_text:
    conversation_with_gpt3p5.run(content)
```
